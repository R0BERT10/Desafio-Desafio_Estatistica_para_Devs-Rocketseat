## Exercício do módulo ##

Com base nas aulas do módulo de Estatística, utilize os métodos da biblioteca Pandas para tratar o dicionário abaixo e:

* Trazer a média das vendas
* Criar um gráfico de barras vertical mostrando o mês de referência e o valor
* Criar um gráfico de linhas mostrando o mês de referência e o valor.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01',
        '2020-02-01',
        '2021-03-01',
        '2022-04-01',
        '2023-05-01',
        '2023-06-01',
        '2020-07-01',
        '2021-08-01',
        '2022-09-01',
        '2023-10-01',
        '2022-11-01',
        '2023-12-01',
        ],
    'valor': [
        400000,
        890000,
        760000,
        430000,
        920000,
        340000,
        800000,
        500000,
        200000,
        900000,
        570000,
        995000,
        ]
}
```

```python
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'], format='%Y-%m-%d')
df_faturamento
```

### Saidas ###

```python
media_faturamneto = df_faturamento.valor.mean()
print(f"O valor medio de vendas e: {media_faturamneto}")
```

```python
df_ordenado = df_faturamento.sort_values(by=['data_ref'])

grafico_vertical = df_ordenado.plot.bar(x="data_ref", y="valor", title=' Faturamento grafico em barras')
grafico_linhas = df_ordenado.plot.line(x="data_ref", y="valor", title=' Faturamento grafico em linhas')
```
